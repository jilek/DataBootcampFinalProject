# Cleaning Data

```python
def ingredient(df):
    recent_chocolate_df["key"] = recent_chocolate_df["REF"].astype(str) + recent_chocolate_df["Specific Bean Origin or Bar Name"]
    df['key']= df ['REF'].astype(str) + df["Bean_Origin_or_Bar_Name"]
    recent_chocolate_df.rename(columns={'Most Memorable Characteristics':'Most_Memorable_Characteristics'}, inplace=True)
    cln_recent_df= recent_chocolate_df[["key", "Ingredients", "Most_Memorable_Characteristics"]]
    df = pd.merge(df,
                     cln_recent_df,
                     on ='key',
                     how ='left')
    df = df.drop(['key'],axis=1)
    
    return df
```

```python
df_chocolate = ingredient(df_chocolate)
df_chocolate.head(20)
```

```python
#Converting String Cocoa_Percent column into Integers
df_chocolate["Cocoa_Percent"] = df_chocolate["Cocoa_Percent"].str.replace('%', '')
df_chocolate['Cocoa_Percent'] = df_chocolate['Cocoa_Percent'].str.replace('.', '')
df_chocolate["Cocoa_Percent"]= df_chocolate["Cocoa_Percent"].astype(float)
df_chocolate["Cocoa_Percent"].value_counts()
```

```python
# Replace empty Ingredients and Most Memorable
df_chocolate['Ingredients'] = df_chocolate['Ingredients'].fillna("Unknown")
df_chocolate['Most_Memorable_Characteristics'] = df_chocolate['Most_Memorable_Characteristics'].fillna("missing")
```

```python
# Replace empty BB_origin
df_chocolate['Broad_Bean_Origin'] = df_chocolate['Broad_Bean_Origin'].fillna(df_chocolate['Bean_Origin_or_Bar_Name'])
```

```python
# Drop null values
df_chocolate.dropna(axis=0, how='any', inplace=True)
```

```python
df_chocolate['Bean_Type'].value_counts()
```

```python
#replace the values in the bean_type column
def replace_beans (bean_type):
    #order of criollo, trinitario, and forastero was decided by seeing the number of times it appears
    if 'Criollo' in bean_type:
        return 'Criollo'
    elif 'Trinitario' in bean_type:
        return 'Trinitario'
    elif 'Forastero' in bean_type:
        return 'Forastero'
    elif 'Nacional' in bean_type:
        return 'Nacional'
    else:
        return bean_type
    
df_chocolate['Bean_Type'] = df_chocolate['Bean_Type'].map(lambda bean_type: replace_beans(bean_type))
```

```python
# Replace the weird spaces with None (Symbolizes no data) 

def repl_space(x):
    if(x is "\xa0"):
        return "missing"
    else:
        return x
       
df_chocolate['Bean_Type'] = df_chocolate['Bean_Type'].apply(repl_space)
df_chocolate.head()
```

```python
def cont_names(df):
    df['continent'] = df['continent'].apply(str).str.replace('Caribbean(DR/Jam/Tri)', "Caribbean",regex=False)\
    .str.replace('South America(SMartin,Pangoa,nacional)', "South America",regex=False)\
    .str.replace('Africa, Caribbean, C. Am.', "Africa",regex=False)
    return df
```

```python
df_chocolate = cont_names(df_chocolate)
df_chocolate['continent'].value_counts()
```

```python
def country_names(df):
    df['Broad_Bean_Origin'] = df['Broad_Bean_Origin'].apply(str).str.replace('Caribbean_rep(DR/Jam/Tri)', "Dominican Republic",regex=False)\
    .str.replace('Peru(SMartin,Pangoa,nacional)', "Peru",regex=False)\
    .str.replace('^Africa, Caribbean_rep, C. Am.$', "Other",regex=True)\
    .str.replace('^Africa_rep$', "Other",regex=True)\
    .str.replace('Venezuela, Caribbean_rep', "Venezuela",regex=True)\
    .str.replace('Caribbean_rep', "Other",regex=True)\
    .str.replace('South America', "Other",regex=True)\
    .str.replace('Other, Africa', "Other",regex=True)\
    .str.replace('Peru(SMartin,Pangoa,nacional)', "Peru",regex=False)
    return df
```

```python
df_chocolate = country_names(df_chocolate)
df_chocolate['Broad_Bean_Origin'].value_counts()
```

```python
# Get latitudes and longitudes
def latlong(df):
    countries_df2=countries_df
#     df['country'] = df['country'].astype('|S80', errors='ignore')
    cln_country_df = countries_df2[["country_code","Broad_Bean_Origin_Country","latitude","longitude"]]
    cln_country_df.head()
    df = pd.merge(df,
                     cln_country_df,
                     on ='Broad_Bean_Origin_Country',
                     how ='left')
    
    return df
    
```

```python
df_chocolate = latlong(df_chocolate)
pd.set_option("display.max_rows", 100)
df_chocolate.head (100)
```

```python
#Set latitude and longitude nulls to 0,0 and country_code to zz
df_chocolate['country_code'] = df_chocolate['country_code'].fillna("ZZ")
df_chocolate['latitude'] = df_chocolate['latitude'].fillna("0.00")
df_chocolate['longitude'] = df_chocolate['longitude'].fillna("0.00")
df_chocolate.isnull().sum()
```
