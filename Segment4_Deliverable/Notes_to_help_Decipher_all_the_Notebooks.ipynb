# Data_Clean.ipynb

1. Add 'Ingredients' and 'Most_Memorable_Characteristics' from webscrape of [flavors_of_cacao](http://flavorsofcacao.com/chocolate_database.html)
2. Reformat 'Cocoa_Percent' column
    + remove '%' char
    + remove '.' char (decimal point) **BUG (72.5 becomes 725)**
    + cast string to float
3. Fix NA values 
    + Change Ingredients=NA to 'Unknown'
    + Change Most_Memorable_Characteristics=NA to 'missing'
    + Call dropna() to catch any other NaN's in df
4. If 'Bean_Type' is a list, pick one of Criollo, Trinitario, or Forastero
5. Change Unicode 'Non-Breaking Space' char to ASCII string 'missing'
6. Fix Country_Location
    + Fix mispellings ('Eucador' -> 'Ecuador', 'Niacragua' -> 'Nicaragua', etc.)
    + Expand acronyms ('U.K.' -> 'United Kingdom', etc.)
    + Change any City entries to their associated Country ('Amsterdam' -> 'Netherlands', etc.)
7. Pick one (most prominent or most likely) country where a list is provided
8. Create a 'Continent' column derived from 'Broad_Bean_Origin' column
9. Clean 'Continent' column
    + Pick one most prominent when a list is given
10. Do the same thing for the 'Broad_Bean_Origin' column
    + Pick one most prominent when a list is given
11. Drop rows where 'Cocoa_Percent' > 100.0 percent (to **fix bug** in #2 above)
12. Create geocoding columns and merge
    + In 'countries' dataframe, use 2-char (ISO 3166-1 alpha-2) values (i.e. 'AD') to match full string (i.e. 'Andorra') in 'Broad_Bean_Origin' column of original dataframe
    + Add 'country_code' (2-char), 'latitude', 'longitude' columns to original dataframe
13. Fix missing / Null values in new geocoding columns
    + If 'country_code' == 'NA' -> 'ZZ'
    + If 'latitude' == 'NA' => '0.00'
    + If 'longitude' == 'NA' => '0.00'
