```python
# copy Broad_Bean_Origin to country, then clean names and join with country lat_long dataframe
def country_names(df):
    colnames = [
            'Company',
            'Bean_Origin_or_Bar_Name',
            'REF',
            'Review_Date',
            'Cocoa_Percent',
            'Company_Location',
            'Rating',
            'Bean_Type',
            'Broad_Bean_Origin'
    ]
    df.columns = colnames
    df['country']= df["Broad_Bean_Origin"]
    df['country'] = df['country'].astype('|S80', errors='ignore')
    df['country'] = df['country'].str.replace('Domincan Republic', 'Dominican Republic')\
    .str.replace('Eucador', 'Ecuador')\
    .str.replace('Amsterdam','Netherlands')\
    .str.replace('Niacragua', 'Nicaragua')\
    .str.replace('Carribean','Caribbean')\
    .str.replace('Sao Tome$', 'Sao Tome & Principe')\
    .str.replace('^Principe$','Sao Tome & Principe')\
    .str.replace('Ivory Coast', "Cote d'Ivorie")\
    .str.replace('Congo', "Republic of Congo")\
    .str.replace('Cost Rica', "Costa Rica")\
    .str.replace('Trinidad$', "Trinidad Tobago")\
    .str.replace('Burma', 'Myanmar')\
    .str.replace('^Tobago', "Trinidad Tobago")\
    .str.replace('Trinidad-Tobago$', "Trinidad Tobago")\
    .str.replace('Trinidad, Tobago', "Trinidad Tobago")\
    .str.replace('St. Lucia', 'Saint Lucia')
    cln_country_df = countries_df[["country_code","country","latitude","longitude"]]
    cln_country_df.head()
    df = pd.merge(df,
                     cln_country_df,
                     on ='country',
                     how ='left')
    
    return df
    
```

```python
chocolate_df = country_names(chocolate_df)
pd.set_option("display.max_rows", 90)
chocolate_df.head (100)
```

```python
def ingredient(df):
    recent_chocolate_df["key"] = recent_chocolate_df["REF"].astype(str) + recent_chocolate_df["Specific Bean Origin or Bar Name"]
    df['key']= df ['REF'].astype(str) + df["Bean_Origin_or_Bar_Name"]
    cln_recent_df= recent_chocolate_df[["key", "Ingredients", "Most Memorable Characteristics"]]
    df = pd.merge(df,
                     cln_recent_df,
                     on ='key',
                     how ='left')
    df.drop(["key"],axis=1)
    
    return df
    
```

```python
chocolate_df = ingredient(chocolate_df)
chocolate_df.head()
```
