```python
# copy Broad_Bean_Origin to country, then clean names and join with country lat_long dataframe
def country_names(df):
    colnames = [
            'Company',
            'Bean_Origin_or_Bar_Name',
            'REF',
            'Review_Date',
            'Cocoa_Percent',
            'Company_Location',
            'Rating',
            'Bean_Type',
            'Broad_Bean_Origin'
    ]
    df.columns = colnames
    df['country']= df["Broad_Bean_Origin"]
    df['country'] = df['country'].astype('|S80', errors='ignore')
    df['country'] = df['country'].str.replace('Domincan Republic', 'Dominican Republic')\
    .str.replace('Eucador', 'Ecuador' ,regex=True)\
    .str.replace('Niacragua', 'Nicaragua',regex=True)\
    .str.replace('Carribean','Other',regex=True)\
    .str.replace('Sao Tome$', 'Sao Tome & Principe',regex=True)\
    .str.replace('^Principe$','Sao Tome & Principe',regex=True)\
    .str.replace('Ivory Coast', "Cote d'Ivorie",regex=True)\
    .str.replace('Congo', "Republic of Congo",regex=True)\
    .str.replace('Cost Rica', "Costa Rica",regex=True)\
    .str.replace('Trinidad$', "Trinidad Tobago",regex=True)\
    .str.replace('Burma', 'Myanmar',regex=True)\
    .str.replace('West Africa', 'Other',regex=True)\
    .str.replace('^Tobago', "Trinidad Tobago",regex=True)\
    .str.replace('Trinidad-Tobago$', "Trinidad Tobago",regex=True)\
    .str.replace('Trinidad, Tobago', "Trinidad Tobago",regex=True)\
    .str.replace('Central and S. America', "Other",regex=True)\
    .str.replace('Colombia, Ecuador', "Colombia",regex=True)\
    .str.replace('South America', "Other",regex=True)\
    .str.replace('Ghana, Domin. Rep', "Ghana",regex=True)\
    .str.replace('Peru, Belize', "Peru",regex=True)\
    .str.replace('Ven.,Ecu.,Peru,Nic.', "Venezuela",regex=True)\
    .str.replace('Venezuela, Dom. Rep.', "Venezuela",regex=True)\
    .str.replace('Ven., Indonesia, Ecuad.', "Venezuela",regex=True)\
    .str.replace('Dom. Rep., Madagascar', "Dominican Republic",regex=True)\
    .str.replace('Venezuela, Caribbean', "Venezuela",regex=True)\
    .str.replace('Ven., Trinidad, Mad.', "Venezuela",regex=True)\
    .str.replace('Venezuela, Java', "Venezuela",regex=True)\
    .str.replace('Peru, Madagascar', "Peru",regex=True)\
    .str.replace('Ven, Bolivia, D.R.', "Venezuela",regex=True)\
    .str.replace('Ven, Trinidad, Ecuador', "Venezuela",regex=True)\
    .str.replace('Costa Rica, Ven', "Costa Rica",regex=True)\
    .str.replace('Trinidad, Ecuador', "Trinidad Tobago",regex=True)\
    .str.replace('Mad., Java, PNG', "Madagascar",regex=True)\
    .str.replace('Ecuador, Costa Rica', "Ecuador",regex=True)\
    .str.replace('Venezuela, Ghana', "Venezuela",regex=True)\
    .str.replace('Gre., PNG, Haw., Haiti, Mad', "Grenada",regex=True)\
    .str.replace('Other(DR/Jam/Tri)', "Other",regex=False)\
    .str.replace('Indonesia, Ghana', "Indonesia",regex=True)\
    .str.replace('Dominican Rep., Bali', "Dominican Republic",regex=True)\
    .str.replace('Africa, Other, C. Am.', "Other",regex=True)\
    .str.replace('Peru, Dom. Rep', "Peru",regex=True)\
    .str.replace('Guat., D.R., Peru, Mad., PNG', "Guatemala",regex=True)\
    .str.replace('Venezuela, Other', "Venezuela",regex=True)\
    .str.replace('Venez,Africa,Brasil,Peru,Mex', "Venezuela",regex=True)\
    .str.replace('DR, Ecuador, Per', "Dominican Republic",regex=True)\
    .str.replace('Peru, Ecuador, Venezuela', "Peru",regex=True)\
    .str.replace('Ecuador, Mad., PNG', "Ecuador",regex=True)\
    .str.replace('Madagascar & Ecuador', "Madagascar",regex=True)\
    .str.replace('Peru, Ecuador', "Peru",regex=True)\
    .str.replace('Peru, Mad., Dom. Rep.', "Peru",regex=True)\
    .str.replace('Venezuela/ Ghana', "Venezuela",regex=True)\
    .str.replace('Dominican Republicu', "Dominican Republic",regex=True)\
    .str.replace('PNG, Vanuatu, Mad', "Papua New Guinea",regex=True)\
    .str.replace('Venezuela, Trinidad Tobago', "Venezuela",regex=True)\
    .str.replace('Peru(SMartin,Pangoa,nacional)', "Peru",regex=False)\
    .str.replace('Other, Africa', "Other",regex=True)\
    .str.replace('Ghana & Madagascar', "Ghana",regex=True)\
    .str.replace('Ghana, Panama, Ecuador', "Ghana",regex=True)\
    .str.replace('^\s*$', "Unknown",regex=True)\
    .str.replace('St. Lucia', 'Saint Lucia',regex=True)
    cln_country_df = countries_df[["country_code","country","latitude","longitude"]]
    cln_country_df.head()
    df = pd.merge(df,
                     cln_country_df,
                     on ='country',
                     how ='left')
    
    return df
    
```

```python
def ingredient(df):
    recent_chocolate_df["key"] = recent_chocolate_df["REF"].astype(str) + recent_chocolate_df["Specific Bean Origin or Bar Name"]
    df['key']= df ['REF'].astype(str) + df["Bean_Origin_or_Bar_Name"]
    recent_chocolate_df.rename(columns={'Most Memorable Characteristics':'Most_Memorable_Characteristics'}, inplace=True)
    cln_recent_df= recent_chocolate_df[["key", "Ingredients", "Most_Memorable_Characteristics"]]
    df = pd.merge(df,
                     cln_recent_df,
                     on ='key',
                     how ='left')
    df = df.drop(['key'],axis=1)
    
    return df
    
```

```python
#replace the values in the bean_type column
def replace_beans (bean_type):
    #order of criollo, trinitario, and forastero was decided by seeing the number of times it appears
    if 'Criollo' in bean_type:
        return 'Criollo'
    elif 'Trinitario' in bean_type:
        return 'Trinitario'
    elif 'Forastero' in bean_type:
        return 'Forastero'
    else:
        return bean_type
    
chocolate_df['Bean_Type'] = chocolate_df['Bean_Type'].map(lambda bean_type: replace_beans(bean_type))
```

```python
chocolate_df = ingredient(chocolate_df)
chocolate_df.head(20)
```
